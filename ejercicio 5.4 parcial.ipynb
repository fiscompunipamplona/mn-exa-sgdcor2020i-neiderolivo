```python
from numpy import cos,sin,pi,zeros
from pylab import *
```

```python
#parte a
def J(m,x):	

    def f(teta):
        return cos(m*teta - x*sin(teta))

    N = 1000
    a = 0.
    b = pi
    h = (b-a)/N
    
    s = f(a) + f(b) + 4*f(b-h)
    for k in range(1,N//2):
        s += 4*f(a + (2*k-1)*h) + 2*f(a+2*k*h)
    
    I = h/3*s/pi
    
    return I

x = linspace(0,20)

plot(x,J(0,x),"r")
plot(x,J(1,x),"y")
plot(x,J(2,x),"g")
legend(['J0','J1','J2'])
```

```python
#parte b
def r(x,y):
    return sqrt(x**2+y**2)

def I(r):
    if (r==0):
        return 1/4
    landa=0.5  #micrometros
    kr=2*pi/landa*r
    return (J(1,kr)/kr)**2
lado=2 #longitud en micrometros
puntos=200 #nmero de puntos de cuadrcula en cada direccin
espaciado=lado/puntos


##calculando la posicion del centro

xcentro=lado/2
ycentro=lado/2

##matriz vaca para almacenar valores

intensidades=zeros([puntos,puntos],float)

##calculo de valores en el vector

for i in range(puntos):
    y=espaciado*i
    for j in range (puntos):
        x=espaciado*j
        dist=r(x-xcentro,y-ycentro)
        intensidades[i,j]=I(dist)   
imshow(intensidades,origin="lower",extent=[0,lado,0,lado],vmax=0.01)
hot()
show()
```
